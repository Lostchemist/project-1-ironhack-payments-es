```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
cash = pd.read_csv('../project_dataset/cash_request.csv')
cash.set_index(['id'], inplace=True)
cash.head()
```

```python
cash.info()
```

```python
cash.replace('NaN', None, inplace=True)
cash.head()
```

```python
cash.count()
```

```python
print("cash - transfer_type")
print(cash["transfer_type"].unique())
print("cash - status")
print(cash["status"].unique())
print("cash - recovery_status")
print(cash["recovery_status"].unique())
print()
```

```python
cash['created_at'] = pd.to_datetime(cash['created_at'])
cash['updated_at'] = pd.to_datetime(cash['updated_at'])
cash['moderated_at'] = pd.to_datetime(cash['moderated_at'],unit='ns')
cash['reimbursement_date'] = pd.to_datetime(cash['reimbursement_date'],unit='ns')
cash['cash_request_received_date'] = pd.to_datetime(cash['cash_request_received_date'],unit='ns')
cash['money_back_date'] = pd.to_datetime(cash['money_back_date'],unit='ns')
cash['send_at'] = pd.to_datetime(cash['send_at'],unit='ns')
cash['reco_creation'] = pd.to_datetime(cash['reco_creation'],unit='ns')
cash['reco_last_update'] = pd.to_datetime(cash['reco_last_update'],unit='ns')
```

```python
cash.info()
```

```python
amount_by_status_sum = sns.barplot(x=cash["amount"], y=cash["status"], estimator=sum, errorbar=None)
amount_by_status_sum.bar_label(amount_by_status_sum.containers[0])
```

```python
amount_by_status_count = sns.barplot(x=cash["amount"], y=cash["status"], estimator='count', errorbar=None)
amount_by_status_count.bar_label(amount_by_status_count.containers[0])
```

```python
amount_by_status_mean = sns.barplot(x=cash["amount"], y=cash["status"], estimator='mean', errorbar=None)
amount_by_status_mean.bar_label(amount_by_status_mean.containers[0])
```

```python
cash['month_created_at'] = cash['created_at'].dt.month
cash['year_created_at'] = cash['created_at'].dt.year
amount_by_status_mean = sns.barplot(x=cash["year_created_at"], y=cash["amount"], estimator='count', errorbar=None)
amount_by_status_mean.bar_label(amount_by_status_mean.containers[0])
```

```python
datos_agrupados = cash.groupby(['year_created_at', 'month_created_at', 'status'])['amount'].sum().reset_index()
sns.lineplot(x='month_created_at', y='amount', hue='status', style='year_created_at', data=datos_agrupados)
```

```python
datos_2020 = cash[cash['created_at'].dt.year == 2020]
datos_agrupados = datos_2020.groupby(['month_created_at', 'status'])['amount'].sum().reset_index()
sns.lineplot(x='month_created_at', y='amount', hue='status', data=datos_agrupados)
```
